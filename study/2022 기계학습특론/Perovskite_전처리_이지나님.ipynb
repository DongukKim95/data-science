```python
# | 가 포함된 숫자 자료 컬럼을 합산 처리, meanFlag = True일시 0인 값은 mean값으로 대체
def calTotal(column, meanFlag) :
    for idx in perovskite_df.index :
        data = perovskite_df.loc[idx][column]
        if (type(data) == str and data.find(' | ') > -1) :
            total = 0
            data = data.replace('nan', '0')            
            for num in data.split(' | ') :
                total += float(num)             
            perovskite_df.loc[idx, column] = round(total, 2)
            
    perovskite_df[column] = perovskite_df[column].astype('float')
    if meanFlag :
        perovskite_df[perovskite_df[column] == 0][column] = round(perovskite_df[column].mean(), 2)
        
# stack grouping 후 나온 key list 결과에 컬럼 값이 포함되지 않을 경우 ETC로 변환        
def stackCheck(column, stack) :    
    for idx in perovskite_df.index :    
        data = perovskite_df.loc[idx][column]
        if (data not in stack) :
            perovskite_df.loc[idx, column] = 'ETC'
```

```python
# 결측치 별도 처리
# Cell
perovskite_df.loc[perovskite_df['Cell_area_measured'].isna(), 'Cell_area_measured'] = 0.1
# Module
perovskite_df.loc[perovskite_df['Module'] == 0, list(perovskite_df.filter(regex='^Module').columns)] = 0
# Substrate

# ETL
## ETL_thickness의 nan, unkown 값 0으로 처리 후, 나머지 조합 값들은 합산 처리, 0으로 남아있는 값들은 평균 처리
perovskite_df.loc[perovskite_df['ETL_thickness'].isna(), 'ETL_thickness'] = 0
perovskite_df.loc[perovskite_df['ETL_thickness'] == 'unkown', 'ETL_thickness'] = 0
calTotal('ETL_thickness', True)    
## ETL Stack 그룹값이 400 이상인 것만 데이터 처리, 그 외에는 ETC로 변환
ETL_stack = perovskite_df.groupby('ETL_stack_sequence').filter(lambda a : len(a) > 400).groupby('ETL_stack_sequence').size()
stackCheck('ETL_stack_sequence', ETL_stack)     
```
