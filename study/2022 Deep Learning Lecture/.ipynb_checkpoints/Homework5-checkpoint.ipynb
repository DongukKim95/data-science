### 5번

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


x_list = [(1, 0), (0, 1), (1, 1)]

for x1, x2 in x_list:
    print(x1, x2)
    net1 = 0.1*x1 + 0.4*x2 + 0.5
    net2 = 0.2*x1 + 0.6*x2 + 0.4
    
    h1 = sigmoid(net1)
    h2 = sigmoid(net2)
    
    net31 = 0.4*h1 + 0.4*h2 + 0.5
    net32 = 0.3*h1 + 0.7*h2 + 0.9
    net33 = 0.5*h1 + 0.1*h2 + 0.1
    
    y1 = np.exp(net31) / (np.exp(net31) + np.exp(net32) + np.exp(net33))
    y2 = np.exp(net32) / (np.exp(net31) + np.exp(net32) + np.exp(net33))
    y3 = np.exp(net33) / (np.exp(net31) + np.exp(net32) + np.exp(net33))
    
    print(f'net1: {net1}')
    print(f'net2: {net2}')
    print(f'h1: {h1}')
    print(f'h2: {h2}')
    print(f'net31: {net31}')
    print(f'net32: {net32}')
    print(f'net33: {net33}')
    print(f'y1: {y1}')
    print(f'y2: {y2}')
    print(f'y3: {y3}')
    print()
```

### 9번

```python
import torch
import numpy
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# hidden layer 2개
class Model2(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 4, bias=True)
        self.linear2 = nn.Linear(4, 4)
        self.linear3 = nn.Linear(4, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.sigmoid(self.linear(x))
        x = self.sigmoid(self.linear2(x))
        x = self.sigmoid(self.linear3(x))
        return x
    
    
# hidden layer 10개
class Model10(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2, 4, bias=True)
        self.linear2 = nn.Linear(4, 4)
        self.linear3 = nn.Linear(4, 4)
        self.linear4 = nn.Linear(4, 4)
        self.linear5 = nn.Linear(4, 4)
        self.linear6 = nn.Linear(4, 4)
        self.linear7 = nn.Linear(4, 4)
        self.linear8 = nn.Linear(4, 4)
        self.linear9 = nn.Linear(4, 4)
        self.linear10 = nn.Linear(4, 4)
        self.linear11 = nn.Linear(4, 1)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self, x):
        x = self.sigmoid(self.linear(x))
        x = self.sigmoid(self.linear2(x))
        x = self.sigmoid(self.linear3(x))
        x = self.sigmoid(self.linear4(x))
        x = self.sigmoid(self.linear5(x))
        x = self.sigmoid(self.linear6(x))
        x = self.sigmoid(self.linear7(x))
        x = self.sigmoid(self.linear8(x))
        x = self.sigmoid(self.linear9(x))
        x = self.sigmoid(self.linear10(x))
        x = self.sigmoid(self.linear11(x))
        return x
    
    
x_train = torch.FloatTensor([[0,0], [0,1], [1,0], [1,1]])
y_train = torch.FloatTensor([[0], [1], [1], [0]])
criterion = torch.nn.BCELoss()
learning_rate = 0.01
nb_epochs = 50000
```

```python
# 2 layer model
model = Model2()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(nb_epochs+1):
    prediction = model(x_train)
    cost = criterion(prediction, y_train)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch % 1000 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, nb_epochs, cost.item()))

        
with torch.no_grad():
    hypothesis = model(x_train)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == y_train).float().mean()
    print('모델의 출력값(Hypothesis): ', hypothesis.detach().cpu().numpy())
    print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
    print('실제값(Y): ', y_train)
    print('정확도(Accuracy): ', accuracy.item())
```

```python
# 10 layer model
model = Model10()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(nb_epochs+1):
    prediction = model(x_train)
    cost = criterion(prediction, y_train)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    
    if epoch % 1000 == 0:
        print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, nb_epochs, cost.item()))

        
with torch.no_grad():
    hypothesis = model(x_train)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == y_train).float().mean()
    print('모델의 출력값(Hypothesis): ', hypothesis.detach().cpu().numpy())
    print('모델의 예측값(Predicted): ', predicted.detach().cpu().numpy())
    print('실제값(Y): ', y_train)
    print('정확도(Accuracy): ', accuracy.item())
```
