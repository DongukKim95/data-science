```python
import numpy as np
```

```python
# 최대 반복 횟수
num_epoch = 5000

# 학습율 (learning_rate)
lr = 0.01

# 초기값
w0 = 1
w1 = 1
w2 = 0

for epoch in range(1, num_epoch+1):    
    w0_plus_t = w0 - lr * (8*w0 + 8*w1 + 12*w2 - 8)
    w1_plus_t = w1 - lr * (8*w0 + 12*w1 + 20*w2 - 10)
    w2_plus_t = w2 - lr * (12*w0 + 20*w1 + 36*w2 - 14)
    
    w0 = w0_plus_t
    w1 = w1_plus_t
    w2 = w2_plus_t
    
    if epoch % 100 == 0:
        print('{0}epoch: w0 = {1:.5f}, w1 = {2:.5f}, w2 = {3:.5f}'.format(epoch, w0, w1, w2))
```
