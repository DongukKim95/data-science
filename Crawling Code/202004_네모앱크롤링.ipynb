```python
# 전체 코드

keywords = seoul_dong.split('\n')  # 법정동 목록
driver = webdriver.Chrome('./chromedriver', options=options)  
driver.get('https://www.nemoapp.kr/')  # 해당하는 url에 접속합니다.
driver.implicitly_wait(10)

# 저장 파일
f = open('nemoapp.csv', 'w')
f.write('법정동명,주거인구,1인가구비율,직장인구\n')

count = 0

for keyword in keywords:
    # 드라이버 변수를 만듭니다. / 크롬을 켭니다.
    try:
        key = '서울시 ' + keyword

        elem = driver.find_element_by_xpath('//*[@id="keyword"]')
        elem.send_keys(key)  # 해당 요소에 문자 입력

        time.sleep(3)

        elem.send_keys(Keys.RETURN) # 엔터키

        driver.implicitly_wait(10)

        driver.find_element_by_xpath('//*[@id="searchPlaceHolder"]/div/div[3]/div[1]/ul/li[2]').click()

        driver.implicitly_wait(10)

        body = driver.find_element_by_xpath('//*[@id="commercialAreaDetailPlaceHolder"]/div[4]/div')
        body.click()

        time.sleep(3)

        body.send_keys(Keys.PAGE_DOWN)

        time.sleep(3)

        con_list = driver.find_elements_by_class_name('box-wrap')

        try:
            f.write(keyword + ',' +
                    con_list[2].text.split('\n')[1][:-1].replace(',','') + ',' + 
                    con_list[2].text.split('\n')[2].split()[2] + ',' + 
                    con_list[3].text.split('\n')[1][:-1].replace(',','') + '\n')
        except:
            print(key)

            for i in range(len(con_list)):
                print(i, con_list[i].text)

            f.write(keyword + ',' + ' ' + ',' + ' ' + ',' + ' \n')

        count += 1
        print('{} / {}'.format(count, len(keywords)))
        driver.implicitly_wait(10)
    except:
        print('error!')
        count += 1
        pass
    
f.close()
```
