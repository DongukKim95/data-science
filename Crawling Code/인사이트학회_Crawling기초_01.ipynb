```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://tv.naver.com/r/"
raw = requests.get(url)  # headers = {'User-agent' : 'Mozilla/5.0'} 방화벽 뚫는 코드
```

```python
# error check
print(raw)
```

```python
# parsing : 일련의 문자열을 유의미한 단어로 구분
html = BeautifulSoup(raw.text, 'html.parser')
```

```python
container = html.select('div.inner')
print(type(container))
print(container)
```

```python
for cont in container:
    title = cont.select_one('dt.title')
    chn = cont.select_one('dd.chn')
    play = cont.select_one('span.hit')
    like = cont.select_one('span.like')
    print(title.text.strip())
    print(chn.text.strip())
    print(play.text.strip())
    print(like.text.strip())
    print()
```

```python
url = 'http://movie.naver.com/movie/running/current.nhn#'
raw = requests.get(url, headers = {'User-agent' : 'Mozilla/5.0'})
html = BeautifulSoup(raw.text, 'html.parser')
container = html.select('dl.lst_dsc')
for cont in container:
    title = 'http://movie.naver.com' + cont.select_one('dt.tit a').attrs['href']
    sub_url = title
    sub_raw = requests.get(sub_url, headers = {'User-agent' : 'Mozilla/5.0'})
    sub_html = BeautifulSoup(sub_raw.text, 'html.parser')
    try:
        cummulative_view = sub_html.select_one('p.count').text.strip()
    except:
        cummulative_view = 'no such infos..'
    print(cummulative_view)
    print(title)
    print()
```
