```python
# 사업자명 매칭 dict 생성
# 1. corp_name과 사업자명이 완전히 일치하는 케이스

# 'corp_name' 값이 2개 이상인 케이스 제외 (dart)
tt1 = pd.DataFrame(data_dart_2['corp_name'].value_counts())
rm_list1 = tt1[tt1['corp_name'] != 1].index.tolist()

# ' 사업장명' 값이 2개 이상인 케이스 제외 (국민연금)
tt2 = pd.DataFrame(data_ins_m[' 사업장명'].value_counts())
rm_list2 = tt2[tt2[' 사업장명'] != 1].index.tolist()

match_dict = {}

for name_1 in tqdm(list(set(data_dart_m.corp_name.unique().tolist()) - set(rm_list1))):
    temp_str = ''
    for name_2 in list(set(data_ins_m[' 사업장명'].unique().tolist()) - set(rm_list2)):
        if name_1 == name_2:
            temp_str = name_2
            
    match_dict[name_1] = temp_str
    
match_dict
```

```python
# 총 1521개 중
# 기업명으로 1대1 매칭되는 케이스 781개
# 기업명으로 1대1 매칭 안되는 케이스 740개
# => 매칭 안되는 케이스는 다른 방법으로 매칭 필요

match_df = pd.DataFrame([match_dict]).T
match_df.columns = ['corp_name']
match_df_1 = match_df[match_df.corp_name != '']
match_df_2 = match_df[match_df.corp_name == '']

match_df_1.shape[0], match_df_2.shape[0]
```

```python
# key값인 사업자명 데이터 정제
# 너무 많은 중복 사업자명이 나오지 않도록 간단하게만 정제

# 공백 제거
pattern_1 = '\s+' 

# 주식회사 문자열 제거
pattern_2 = '주식회사'

# 괄호안 문자열 제거 -> 가장 짧은 단어 매칭
pattern_3 = '\(.*?\)'

# 괄호안 문자열 제거 2 -> 가장 짧은 단어 매칭
pattern_4 = '\（.+?\）'

# / 문자 뒤 문자열 제거
pattern_5 = '\/.*'

# - 문자 뒤 문자열 제거
pattern_6 = '\-.*'

def text_preprocess(text):
    text = re.sub(pattern_1, '', text)
    text = re.sub(pattern_2, '', text)
    text = re.sub(pattern_3, '', text)
    text = re.sub(pattern_4, '', text)
    #text = re.sub(pattern_5, '', text)
    #text = re.sub(pattern_6, '', text)
    return text
```

```python
data_dart_m['key'] = data_dart_m.stock_name.apply(text_preprocess)
data_ins_m['key'] = data_ins_m[' 사업장명'].apply(text_preprocess)

data_dart_m.key, data_ins_m.key
```

```python
# 2. stock_name과 정제된 사업자명이 완전히 일치하는 케이스

# 1번에서 불일치 케이스만 선별
stock_name_list = data_dart_m.loc[data_dart_m.corp_name.isin(match_df_2.index.tolist()), 'key'].unique().tolist()

# key값이 2개 이상인 케이스 제외 (dart)
tt1 = pd.DataFrame(data_dart_2['stock_name'].apply(text_preprocess).value_counts())
rm_list1 = tt1[tt1.stock_name != 1].index.tolist()

# key값이 2개 이상인 케이스 제외 (국민연금)
tt2 = pd.DataFrame(data_ins_m['key'].value_counts())
rm_list2 = tt2[tt2.key != 1].index.tolist()

match_dict2 = {}

for name_1 in tqdm(list(set(stock_name_list) - set(rm_list1))):
    temp_str = ''
    for name_2 in list(set(data_ins_m['key'].unique().tolist()) - set(rm_list2)):
        if name_1 == name_2:
            temp_str = name_2
            
    match_dict2[name_1] = temp_str
    
match_dict2
```

```python
# 기업명으로 1대1 매칭 안되는 케이스 총 738개 중
# key값으로 매칭되는 케이스 342개
# key값으로 매칭 안되는 케이스 396개

match_df2 = pd.DataFrame([match_dict2]).T
match_df2.columns = ['key']
match_df2_1 = match_df2[match_df2.key != '']
match_df2_2 = match_df2[match_df2.key == '']

match_df2_1.shape[0], match_df2_2.shape[0]
```

```python
# DART & 국민연금 데이터 통합
# 기업명으로 1대1 매칭되는 케이스 781개 +
# key값으로 매칭되는 케이스 342개
# => 총 1123개 기업

# merge1 -> corp_name 기준
data_m_1 = data_dart_m[data_dart_m.corp_name.isin(match_df_1.index.tolist())].merge(
    data_ins_m[data_ins_m[' 사업장명'].isin(match_df_1.index.tolist())], left_on='corp_name', right_on=' 사업장명', how='inner')
data_m_1
```

```python
# merge2 -> key 기준
data_m_2 = data_dart_m[data_dart_m.key.isin(match_df2_1.index.tolist())].merge(
    data_ins_m[data_ins_m.key.isin(match_df2_1.index.tolist())], on='key', how='inner')
data_m_2
```

```python
# merge1 & merge2

data_m = pd.concat([data_m_1, data_m_2])
data_m.reset_index(inplace=True, drop=True)
data_m
```
