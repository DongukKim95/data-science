```python
# 2014년 ,2018년, 총계(5년치 합) 컬럼 삭제 

total1 = total1.drop(total1[['2014년', '2018년', '총계']], axis = 1)
total2 = total2.drop('총계', axis = 1)
```

```python
경기.head()
```

```python
total = pd.concat([total1, total2], axis = 0, ignore_index = True)
```

```python
# 결측치 0으로 처리

for i in range(6,9):
    for j in range(0, len(total)):
        if total.iloc[j][i] == '-':
            total.iloc[j][i] = 0
```

```python
# 방문객 수가 string으로 처리되어 있으므로 float 형식으로 변환

for i in range(6, 9):
    for j in range(0, len(total)):
        if type(total.iloc[j][i]) == str:
            total.iloc[j][i] = total.iloc[j][i].replace(',', '')
            total.iloc[j][i] = float(total.iloc[j][i])
        elif type(total.iloc[j][i]) == int:
            total.iloc[j][i] = float(total.iloc[j][i])
```

```python
# 관광객 수가 집계되지 않은 컬럼?

len(total.loc[(total['2015년']==0) |(total['2016년']==0)|(total['2017년']==0), : ])
```

```python
# 집계 안된 년도가 뒤죽박죽 -> 일관된 처리 어려우므로 모두 삭제

total = total.loc[(total['2015년']!=0) & (total['2016년']!=0) & (total['2017년']!=0), : ]
```

```python
len(total)
```

```python
# 시군별 방문객 수 합계 테이블
totalsum = total.pivot_table(index = ['시도', '군구'], aggfunc = 'sum',
                values = ['2015년','2016년','2017년'])
```

```python
totalsum.reset_index(inplace = True)
```

```python
# 지역 별 관광지 수를 나타내는 컬럼 생성
count = pd.DataFrame(pd.value_counts(total['군구']))
count.reset_index(inplace = True)
count.columns = ['군구', '관광지수']
totalsum = pd.merge(totalsum, count, on = '군구')
```

```python
# 3년 간 관광객 수 총합 / 관광지 수 (관광지가 많은 도시)

totalsum['평균관광객수'] = (totalsum['2015년']+totalsum['2016년']+totalsum['2017년'])/totalsum['관광지수']
```

```python
totalsum.loc[totalsum.시도 == '경기도', '시도'] = '경기'
totalsum.loc[totalsum.시도 == '강원도', '시도'] = '강원'
totalsum.loc[totalsum.시도 == '충청북도', '시도'] = '충북'
totalsum.loc[totalsum.시도 == '충청남도', '시도'] = '충남'
totalsum.loc[totalsum.시도 == '경상북도', '시도'] = '경북'
totalsum.loc[totalsum.시도 == '경상남도', '시도'] = '경남'
totalsum.loc[totalsum.시도 == '전라북도', '시도'] = '전북'
totalsum.loc[totalsum.시도 == '전라남도', '시도'] = '전남'
totalsum.loc[totalsum.시도 == '제주특별자치도', '시도'] = '제주'
totalsum.loc[totalsum.시도 == '서울특별시', '시도'] = '서울'
totalsum.loc[totalsum.시도 == '울산광역시', '시도'] = '울산'
totalsum.loc[totalsum.시도 == '대구광역시', '시도'] = '대구'
totalsum.loc[totalsum.시도 == '부산광역시', '시도'] = '부산'
totalsum.loc[totalsum.시도 == '세종특별시', '시도'] = '세종'
totalsum.loc[totalsum.시도 == '광주광역시', '시도'] = '광주'
totalsum.loc[totalsum.시도 == '대전광역시', '시도'] = '대전'
totalsum.loc[totalsum.시도 == '인천광역시', '시도'] = '인천'
```

```python
aa = pd.merge(totalsum, data)
```

```python
비인기관광지 = aa.loc[aa['평균관광객수'] < aa['평균관광객수'].mean(), :].sort_values(by = '평균관광객수')
인기관광지 = aa.loc[aa['평균관광객수'] >= aa['평균관광객수'].mean(), :].sort_values(by = '평균관광객수')
```
